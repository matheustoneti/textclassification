```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow import keras
```

```python
train = pd.read_csv('/content/Poem_classification - train_data.csv')
```

```python
train.head(10)
```

```python
train['Genre'].value_counts()
```

```python
train = train.replace({'Music': 0, 'Death' : 1, 'Environment': 2, 'Affection': 3})
```

```python
train.isnull().sum()
train.dropna(inplace=True)
```

```python
train.head(10)
```

```python
poemas = train['Poem']
target = train['Genre']
```

```python
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(poemas)

word_index = tokenizer.word_index

poemas = tokenizer.texts_to_sequences(poemas)
poemas_padded = pad_sequences(poemas, maxlen=200, padding='post', truncating='post')
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(10000,16,input_length=200))
model.add(tf.keras.layers.GlobalAveragePooling1D())
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(50,activation='relu'))
model.add(tf.keras.layers.Dense(25,activation='relu'))
model.add(tf.keras.layers.Dense(4,activation='sigmoid'))
opt = tf.keras.optimizers.Adam()
model.compile(loss='sparse_categorical_crossentropy',optimizer=opt,metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(poemas_padded,target, epochs=100, verbose=2)
```
